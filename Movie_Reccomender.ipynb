```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df['tags'] = df['genre'] + df['overview']
```

```python
new_df = df[['id','title','overview','genre','tags']]
```

```python
new_df = new_df.drop(columns =['genre','overview'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english') #Defining cv with class CountVectorizer with max_features = 10000
                                                                #and stop_words = 'english' (max_features  = it is For example, if
                                                                #you set max_features to 1000,
                                                                # then only the 1000 most common terms will be included in the vocabulary)
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
'''
.values: This extracts the actual values from the "tags" column as a NumPy array.
         This is because machine learning algorithms typically work better with numerical data.
.astype('U'): This converts the data type of the values in the array to a Unicode string type ('U').
              This ensures proper handling of text data.

              (By default, CountVectorizer creates a sparse matrix,
              which is memory-efficient for storing text data with many zeros (words not appearing in a document).)
.toarray() : Converts the sparse matrix returned by fit_transform into a dense NumPy array (vec).
             This array will have the same dimensions as the sparse matrix, but all values will be explicitly stored, even zeros.
'''
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
'''
  The cosine_similarity function from sklearn.metrics.pairwise in scikit-learn calculates the cosine similarity
  between samples in two datasets. It's a metric used to compare the similarity between data points represented as vectors.
'''
```

```python
sim =  cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['title']=='The Shawshank Redemption']
```

```python
dist = sorted(list(enumerate(sim[0])),reverse=True,key = lambda vec:vec[1])
'''
  Now,
     Enumerate(sim[0]) This function creates an iterator that pairs the index (position) of each element in sim[0] with
     its corresponding value. It essentially turns sim[0] (a list of similarity scores) into a list of tuples, where each tuple has two elements:
     The first element is the index (position) of a movie in the original sim array.
     The second element is the cosine similarity score between that movie and the target movie (represented by sim[0]).
  vec[1] accesses the second element of the tuple which is the cosine similarity score
  bascically sorted sorts based on the cosine similarity scores
'''
```

```python
dist
```

```python
for i in dist[1:6]:
  print(new_df.iloc[i[0]].title)
```

```python
def recommend(movies):
  index = new_df[new_df['title'] == movies].index[0]
  distances = sorted(list(enumerate(sim[index])),reverse=True,key = lambda vec:vec[1])
  for i in distances[0:10]:
    print(new_df.iloc[i[0]].title)
```

```python
mov = input()
recommend(mov)
```
